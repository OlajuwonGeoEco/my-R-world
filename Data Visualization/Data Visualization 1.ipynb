##Plotting in R

```r
df<-(read.csv(file.choose(),header=T))
```

```r
library("tidyverse")
```

```r
head(df)
```

```r
##creating a scatter plot (continous x continous)
```

```r
ggplot(df)+
geom_point(aes(chol,age, color= sex)) #basic scatter plot
```

```r
ggplot(df)+
geom_point(aes(chol,age, color= sex, size=age))  #exploring more explanatory approach
```

```r
ggplot(df)+
geom_point(aes(chol,age, color= sex, size=age), alpha=0.8)+theme_classic()  #exploring more explanatory approach
```

```r
ggplot(df)+
geom_point(aes(chol,age,color=age),size=4, alpha=0.8)+theme_classic() +facet_wrap(~sex,nrow=1)
```

##Investigating trends quickly

```r
ggplot(df2)+
geom_point(aes(bmi,charges,color=sex))+theme_classic() 
#data looks all scruffy and we can't find a pattern. Lets add some smoothing
```

```r
ggplot(df2)+
geom_smooth(aes(bmi,charges,linetype=sex), se=F)+
theme_classic() 
```

###visit https://ggplot2.tidyverse.org/index.html for titles, labesls, axis, and theme modification

```r
ggplot(df2)+
geom_boxplot(aes(region, charges,fill=region))+theme_classic()  #boxplot for continous x categories
```

```r
ggplot(df2)+
geom_col(aes(region, charges, fill=region))+theme_classic()  #barplot for continous x categories #what does this display?
```

```r
ggplot(df2)+
geom_col(aes(region, charges, fill=smoker))+theme_classic()+coord_flip()  
#barplot for continous x categories #what does this display?
#I promise these plots look better in R studio, or when you resize

ggplot(df2)+
geom_col(aes(region, charges, fill=smoker),position="dodge")+theme_classic()+coord_flip() 
```

```r
ggplot(df2)+
geom_col(aes(region, charges, fill=smoker),position="identity")+theme_classic()+coord_polar() 
```
