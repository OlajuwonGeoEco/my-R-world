```r
library(tidyverse)
```

```r
df<-read.csv(file.choose(),header=TRUE)
```

```r
head(df)
```

charges= Predictor variable
age= dependent variable

```r
lm(age~bmi,data=df)
```

```r
model<-lm(age~bmi,data=df)
```

```r
summary(model)
```

###Whilst the t-statistics and p-values indicate a significant association, the strength of the association is captured by the R2 value.

The F-statistic and associated p-value indicates whether the model as a whole is significant. 

```r
confint(model)
```

##Assumptions Testing

```r
plot(model,which=1)
```

```r
par(mfrow = c(1, 2)) # This code put two plots in the same window
hist(model$residuals) # Histogram of residuals
plot(model, which = 2) # Quantile plot
```

##Visualization 

```r
plot(age ~ bmi, data = df, xlab = "BMI", ylab = "Age", pch = 16)
abline(model, col = "red")
```
